Пусть заданы векторы $\vec{u} = [1, 3, 3]^\top$ и $\vec{v} = [3, -2, 0]^\top$ в $\mathbb{R}^{3}$. Найти уравнение плоскости и визуализировать (с векторами).

Реализовать 2 функции:

1я функция создает данные (1 балл за создание):+

принимает 3 аргумента - количество точек, смещение и разброс
возвращает 2 аргумента X и y - numpy массивы

Где n - количество примеров. Мы будем использовать n = 150
X.shape = n, 2
y.shape = n

Вывод Х предоставлен.

```python
import numpy as np
import random

def createData(amm, bias, variance):
    X = np.array([np.ones(amm), np.arange(0, amm)]).T
    y = []
    for i in range(amm):
        y.append(np.random.normal(X[i][1] + bias, variance))
    y = np.array(y)
    return X, y

# точка данных
data_dots_ammount = 180
# смещение
bias = 20
# разброс
variance = 10
X, y = createData(data_dots_ammount, bias, variance)
```

2я функция - реализация градиентного спуска:

Принимает на вход X, y, веса, альфа, количество точек, количество итераций

Веса можно инициализировать случайными значениями (одномерных массив numpy) из диапазона 0..1 или единицами. Всего 2 значения - это прогнозируемое (на основе данных) смещение и наклон ($ax + b$, b и а соответсвенно).

Пройдясь по всем итерациям:

1. Находим вектор прогнозов $y = X^TW$ (где X, W - данные и веса)
2. Обновляем веса $W = w - \alpha (X.T, y)$ (альфа пологаем 0.03)

Возвращаем параметр весов (2 элемента), далее нужны для построения прогноза

Сохраняем изменение сумму квадратов разностей между прогнозом и реальным значением, делим на 2 * кол-во элементов

```python
import numpy as np
import random

alpha = 0.0000003
n = 180
iteration = 50000
w = np.array([random.uniform(0, 1) for i in range(2)])

def gradientDescent(X, y, weights, alpha, n, iteration):
    mse = []
    for i in range(iteration):
        y_predict = X @ weights
        loss = y_predict - y
        weights = weights - alpha * (X.T @ loss )
        
        mse.append(np.sum(loss ** 2) / (2*n))
        #print(f'iteration {i} |mse: {mse[i]}')
    mse = np.array(mse)
    return weights, mse


w, losses = gradientDescent(X, y, w, alpha, n, iteration)
w
```
