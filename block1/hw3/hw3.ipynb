Задача №1 (1 балл)+

Найти номер клиента и количество его заказов (таблица orders) по убыванию 

Подсказка: 

Order by название_столбца_по_которому_убываем desc;

```python
# решил дополнительно упорядочить по номеру клиента для красоты
df = pd.read_sql(
'''
SELECT customers.customerNumber as client_num, 
    COUNT(orders.orderNumber) as count_orders
FROM customers
INNER JOIN orders
ON customers.customerNumber = orders.customerNumber
GROUP BY customers.customerNumber
ORDER BY count_orders DESC, client_num;
''', connection)
df.head(10)
```

Задача № 5 (3 балла)+

Создать представление с полями:

Название колонки | Источник иформации
--- | ---
Номер заказа | столбец orderNumber таблицы orders
Название продукта | столбец productName таблицы products
Код продукта | столбец productCode таблицы products
Количество заказов | столбец quantityOrdered таблицы orderdetails
Цена за один | столбец priceEach таблицы orderdetails
Цена покупки | столбец buyPrice таблицы products
Итоговая сумма | результат перемножения цены покупки * кол-во заказанных
Прибыль | результат перемножения (закупочной цены (buy Price из products) минус цена покупки) * кол-во заказанных
Заказчик | столбец customerName таблицы customers
Имя представителя | конкатенация столбцев contactLastName и contactFirstName таблицы customers
адрес | столбец addressLine1 таблицы customers

Отобразить результат вызоав представления

Из полученного представления найти заказчиков и принесенную прибыль, (отсортировать по убыванию суммарной прибыли по всем), отобразить

```python
# сначала написал запрос и проверил себя, правильно ли составил
df = pd.read_sql(
'''
SELECT orders.orderNumber as Order_num,
    products.productName as Product_name,
    products.productCode as Product_code,
    orderdetails.quantityOrdered as Count_orders,
    orderdetails.priceEach as Price_one,
    products.buyPrice as Price_of_buy,
    orderdetails.priceEach * orderdetails.quantityOrdered as Result_sum,
    (orderdetails.priceEach - products.buyPrice) * orderdetails.quantityOrdered as Profit,
    customers.customerName as Customer_name,
    CONCAT(customers.contactLastName,' ',customers.contactFirstName) as Contact_Name,
    customers.addressLine1 as Address,
    orders.status as Order_status
FROM orders
INNER JOIN customers
ON orders.customerNumber =  customers.customerNumber 
INNER JOIN orderdetails
ON orders.orderNumber = orderdetails.orderNumber
INNER JOIN products
ON orderdetails.productCode = products.productCode;
''',  connection)
df.head(10)
```

```python
# затем сделал представление из этого запроса (для новго запуска нужно дропать таблицу)
connection.execute(
'''
CREATE VIEW fith_task as
SELECT orders.orderNumber as Order_num,
    products.productName as Product_name,
    products.productCode as Product_code,
    orderdetails.quantityOrdered as Count_orders,
    orderdetails.priceEach as Price_one,
    products.buyPrice as Price_of_buy,
    orderdetails.priceEach * orderdetails.quantityOrdered as Result_sum,
    (orderdetails.priceEach - products.buyPrice) * orderdetails.quantityOrdered as Profit,
    customers.customerName as Customer_name,
    CONCAT(customers.contactLastName,' ',customers.contactFirstName) as Contact_Name,
    customers.addressLine1 as Address,
    orders.status as Order_status
FROM orders
INNER JOIN customers
ON orders.customerNumber =  customers.customerNumber 
INNER JOIN orderdetails
ON orders.orderNumber = orderdetails.orderNumber
INNER JOIN products
ON orderdetails.productCode = products.productCode;
''')
```

```python
# отправим запрос на подтверждение результата
df = pd.read_sql('''
SELECT * from fith_task''', connection)
df.head(10)
```

Из полученного представления найти заказчиков и принесенную прибыль, (отсортировать по убыванию суммарной прибыли по всем)

```python
df = pd.read_sql(
'''
SELECT Customer_name, SUM(Profit) as Summary_profit
FROM fith_task
WHERE Order_status = "Shipped"
GROUP BY Customer_name
ORDER BY Summary_profit DESC
''', connection)
df.head(10)
```
